```python
xn = np.arange(0.0,1.0,step = 1/float(len(nresults["factor_error"])))
xngd = np.arange(0.0,1.0,step = 1/float(len(ngdresults["factor_error"])))
xngde = np.arange(0.0,1.0,step = 1/float(len(ngderesults["factor_error"])))
```

```python
plt.figure(figsize=(12,10))
plt.plot(np.log(xn), nresults["factor_error"], label = "ALS")
plt.plot(np.log(xngd), ngdresults["factor_error"], label = "GD")
plt.plot(np.log(xngde), ngderesults["factor_error"], label = "GD-ELS")
plt.legend()
plt.title("Factor errors for a 20 x 20 x 20 tensor with rank 5")
plt.xlabel("Log time (s)")
plt.ylabel("Sum factor error")
plt.savefig("factor_errors.png", dpi=350)
```
